```python
print("hello word")
# scaping character
print("hello \n word")  # \n is a new line  
print("hello \t word")  # \t is a tab
print("hello \\ word")  # \\ is a backslash
print("hello ' word")  # ' is a single quote
print("hello \" word")  # " is a double quote
print("hello ` word")  # ` is a backtick

# comments single line
# this is a comment, python will ignore it
print("hello word")  # this is a comment, python will not ignore it
# comments multi line
"""
This is a comment
written in
more than just one line
"""
print("Hello, World!")
# separator
print("hello", "word")  # separator
print("hello" + "word")  # separator
print("aman",6,6,6,sep="-",end="009\n")
print("aman")

```
